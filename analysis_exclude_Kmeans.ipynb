#### u-test on growth rate

```python
from scipy.stats import mannwhitneyu

# Calculate the monthly changes in total value and total net value
monthly_total_value_diff = monthly_total_value.diff()
monthly_total_net_value_diff = monthly_total_net_value.diff()

# Create an empty list to store the results
significant_changes = []

# Perform the Mann-Whitney U test for every two consecutive months
for i in range(len(monthly_total_value_diff) - 1):
    month1 = monthly_total_value_diff.index[i]
    month2 = monthly_total_value_diff.index[i + 1]
    
    # Perform the Mann-Whitney U test for total value
    _, p_value_value = mannwhitneyu(sales_analysis_table_delivered_only[sales_analysis_table_delivered_only['year_month'] == month1]['total_value'],
                                    sales_analysis_table_delivered_only[sales_analysis_table_delivered_only['year_month'] == month2]['total_value'])
    
    # Perform the Mann-Whitney U test for total net value
    _, p_value_net_value = mannwhitneyu(sales_analysis_table_delivered_only[sales_analysis_table_delivered_only['year_month'] == month1]['total_net_value'],
                                        sales_analysis_table_delivered_only[sales_analysis_table_delivered_only['year_month'] == month2]['total_net_value'])
    
    if p_value_value < 0.05 or p_value_net_value < 0.05:
        significant_changes.append((month1, month2))

print("Significant changes:")
for change in significant_changes:
    print(f"{change[0]} -> {change[1]}")

```

```python
rfmTable = orders.groupby('customer').agg({'order_date': lambda x: (NOW - x.max()).days, # Recency
                                        'order_id': lambda x: len(x),      # Frequency
                                        'grand_total': lambda x: x.sum()}) # Monetary Value

rfmTable['order_date'] = rfmTable['order_date'].astype(int)
rfmTable.rename(columns={'order_date': 'recency', 
                         'order_id': 'frequency', 
                         'grand_total': 'monetary_value'}, inplace=True)
```

```python
quantiles = rfmTable.quantile(q=[0.25,0.5,0.75])
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
rfmSegmentation = rfmTable
```

```python
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def RClass(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FMClass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
rfmSegmentation['R_Quartile'] = rfmSegmentation['recency'].apply(RClass, args=('recency',quantiles,))
rfmSegmentation['F_Quartile'] = rfmSegmentation['frequency'].apply(FMClass, args=('frequency',quantiles,))
rfmSegmentation['M_Quartile'] = rfmSegmentation['monetary_value'].apply(FMClass, args=('monetary_value',quantiles,))
```

```python
rfmSegmentation['RFMClass'] = rfmSegmentation.R_Quartile.map(str) \
                            + rfmSegmentation.F_Quartile.map(str) \
                            + rfmSegmentation.M_Quartile.map(str)
```

```python
# Define the function cust_score_title_lkup
def cust_score_title_lkup(rfm_score):
    if rfm_score >= 444:
        return 'Platinum Customer'
    elif 433 <= rfm_score < 444:
        return 'Very Loyal'
    elif 421 <= rfm_score < 433:
        return 'Becoming Loyal'
    elif 344 <= rfm_score < 421:
        return 'Recent Customer'
    elif 323 <= rfm_score < 344:
        return 'Potential Customer'
    elif 311 <= rfm_score < 323:
        return 'Late Bloomer'
    elif 224 <= rfm_score < 311:
        return 'Losing Customer'
    elif 212 <= rfm_score < 224:
        return 'High Risk Customer'
    elif 124 <= rfm_score < 212:
        return 'Almost Lost Customer'
    elif 112 <= rfm_score < 124:
        return 'Evasive Customer'
    else:
        return 'Lost Customer'

# Apply the function to the RFMClass column
rfmSegmentation['RFMClass'] = rfmSegmentation['RFMClass'].astype(int)

rfmSegmentation['RFMClass_Title'] = rfmSegmentation['RFMClass'].apply(cust_score_title_lkup)
```
